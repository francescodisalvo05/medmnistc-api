```python
from medmnistc.visualizer import Visualizer

from PIL import Image
```

### Visualize one severity per corruption

```python
vis = Visualizer(
    medmnist_path='/mnt/data/datasets/medmnist',
    medmnistc_path='/mnt/data/datasets/medmnistc',
    output_path='images' # where to store the output images
)

dataset = "breastmnist"

vis.plot_one_severity(dataset_name = dataset, severity= 3)
```

```python
Image.open('images/one_severity/breastmnist_id2_sev3.png')
```

### Visualize all corruptions and severities
